```python
import json
import requests
#Post/sensor_api
URL = 'http://192.168.15.5:5000'

```

```python
body_sensor = {
    "tipo": "temperatura 12ttt", 
    "dados": "1233.44"
}
```

```python
headers_sensor = {
    'Content-Type':'application/json'
}
```

```python
post_sensor = requests.request('POST', URL + '/sensor_api', json=body_sensor, headers=headers_sensor)
```

```python
print(post_sensor.status_code)
```

```python
post_sensor.json()
```
